1. Importing data set using Pandas

```python
import pandas as pd
cov = pd.read_csv("https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv")
cov
```

b. Find data types of columns

```python
cov.dtypes
```

c. Find info & describe of data in dataframe

```python
cov.info()
cov.describe(include='all') #Or cov.describe()
```

<h3>3. Low Level Data Understanding:</h3>
a. Find count of unique values in location column

```python
cov["location"].nunique()
```

b. Find which continent has maximum frequency using value counts

```python
cov["continent"].value_counts()
```

c. Find maximum & mean value in 'total_cases'

```python
print(cov["total_cases"].max())
print(cov["total_cases"].mean())
```

d. Find 25%, 50% & 75% quartile value in 'total_deaths'

```python
cov["total_deaths"].quantile([0.25, 0.5, 0.75])
```

e. Find which continent has maximum 'human_development_index'

f. Find which continent has minimum 'gdp_per_capita'

```python
cov.groupby("continent").agg({"human_development_index":"max", "gdp_per_capita":"min"})
```

4. Filter the DataFrame with only these columns ['continent', 'location', 'date', 'total_cases', 'total_deaths', 'gdp_per_capita', 'human_development_index'] and update the data frame.

```python
cov = pd.DataFrame(cov[['continent', 'location', 'date', 'total_cases', 'total_deaths', 'gdp_per_capita', 'human_development_index']])
cov
```

<h3>5. Data Cleaning:</h3>
a. Remove all duplicates observations

```python
cov.duplicated().sum()
```

```python
cov.drop_duplicates()  #However no duplicates are present as seen above
```

b. Find missing values in all columns

```python
cov.isnull().sum()
```

c. Remove all observations where continent column value is missing
<br>Tip : using subset parameter in dropna

```python
cov = cov.dropna(subset=['continent'])
cov
```

d. Fill all missing values with 0

```python
cov = cov.fillna(0)
cov
```

<h3>6. Date time format:</h3>
a. Convert date column in datetime format using pandas.to_datetime

```python
cov["date"] = pd.to_datetime(cov['date'])
```

b. Create new column month after extracting month data from date column

```python
cov["month"] = pd.DatetimeIndex(cov['date']).month
```

<h3>7. Data Aggregation:</h3>
a. Find max value in all columns using groupby function on 'continent' column
<br>Tip: use reset_index() after applying groupby

b. Store the result in a new dataframe named 'df_groupby'
<br>(Use df_groupby dataframe for all further analysis)

```python
df_groupby = cov.groupby("continent").mean()
df_groupby = df_groupby.reset_index()
df_groupby
```

<h3>8. Feature Engineering :</h3>
a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'

```python
df_groupby['total_deaths_to_total_cases'] = df_groupby["total_deaths"]/df_groupby["total_cases"]
```

<h3>9. Data Visualization:</h3>
a. Perform univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot

```python
import seaborn as sns
sns.distplot(df_groupby['gdp_per_capita'], kde=False)
```

b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'

```python
sns.jointplot(data=df_groupby, x='total_cases', y='gdp_per_capita', kind="scatter")
```

c. Plot pairplot on df_groupby dataset

```python
sns.pairplot(data=df_groupby)
```

d. Plot a bar plot of 'continent' column with 'total_cases'
<br>Tip : using kind='bar' in seaborn catplot

```python
sns.catplot(data=df_groupby, x="continent", y="total_cases", kind='bar', aspect=2)
```
